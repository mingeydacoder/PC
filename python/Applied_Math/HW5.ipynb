```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

x = np.linspace(0,20,100)

for n in range(2):
    y = sp.jn(n,x)
    plt.plot(x,y,label=r'$J_ %s(x)$'%n)
plt.axhline(0,color='green',label='_nolegend_')
plt.grid()
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$J_m (x)$')
plt.title('Bessel Functions')

```
